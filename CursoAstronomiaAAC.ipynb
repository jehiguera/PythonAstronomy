# **Ejemplo de tratamiento de imagenes en dos dimensiones**

Aunque el paquete nddata admite cualquier tipo de datos, este ejemplo  se centrará en el uso de nddata para imágenes bidimensionales. Para comenzar, construiremos una imagen bidimensional con algunas fuentes, algo de ruido gaussiano y un "rayo cósmico" que luego se enmascara.

Primero se construye una imagen bidimensional con algunas fuentes, algo de ruido gaussiano y un "rayo cósmico":

https://en.wikipedia.org/wiki/Gaussian_function

```python
import numpy as np
```

```python
from astropy.modeling.models import Gaussian2D
```

```python
y, x = np.mgrid[0:500, 0:600]
```

https://docs.astropy.org/en/stable/api/astropy.modeling.functional_models.Gaussian2D.html

Parametros de la funcion Gaussian2D(amplitude=1, x_mean=0, y_mean=0, x_stddev=None, y_stddev=None, theta=None, cov_matrix=None, **kwargs)

Amplitude (peak value) of the Gaussian.
Mean of the Gaussian in x.
Mean of the Gaussian in y.
Standard deviation of the Gaussian in x before rotating by theta. 
Standard deviation of the Gaussian in y before rotating by theta. 
theta: Rotation angle (value in radians).

```python
data = (Gaussian2D(1, 150, 100, 20, 10, theta=0.5)(x, y) +
...         Gaussian2D(0.5, 400, 300, 8, 12, theta=1.2)(x,y) +
...         Gaussian2D(0.75, 250, 400, 5, 7, theta=0.23)(x,y) +
...         Gaussian2D(0.9, 525, 150, 3, 3)(x,y) +
...         Gaussian2D(0.6, 200, 225, 3, 3)(x,y))
```

```python
data += 0.01 * np.random.randn(500, 600)
```

```python
cosmic_ray_value = 0.997
```

```python
data[100, 300:310] = cosmic_ray_value
```

Esta imagen tiene una gran "galaxia" en la parte inferior izquierda y hay un "rayo cósmico" en una línea horizontal en la parte inferior central de la imagen:

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(data, origin='lower')
```

El "rayo cósmico" se puede enmascarar en esta imagen de prueba, así:

```python
mask = (data == cosmic_ray_value)
```

# **Calcular orbita de transferencia**

Existen diferentes tipos de transferencias orbitales y métodos que pueden ser utilizados por aquellos satélites de propulsión impulsiva. Estas son la transferencia bi-impulsiva y transferencia de Hohmann, la transferencia triimpulsiva, el método de aproximación por cónicas, el método del problema de Lambert y la asistencia gravitacional o flyby. Donde con la salvedad del flyby, cuya misión es acelerar el satélite sin que sea requerido consumir combustible, el resto de trasferencias dan solución al problema del cálculo de la órbita de transferencia de una posición a otra


La órbita de transferencia entre dos puntos en el espacio solo depende de esos dos puntos y del tiempo que se tarda en pasar de uno a otro.

http://www.aero.us.es/move/files/t2_parte2.pdf

http://www.aero.us.es/move/files1617/Tema7_1617.pdf

 Usamos algoritmos disponibles en poliastro.iod para resolver esto, ¡pero trabajar con poliastro.maneuvers es aún más fácil!

Paso1: Crear las orbitas de cada uno de los planetas en la fecha de salida de la tierra y llegada a Marte:

```python
earth = Ephem.from_body(Earth, time_range(date_launch, end=date_arrival))
mars = Ephem.from_body(Mars, time_range(date_launch, end=date_arrival))
```

Resolver para órbitas de salida y orbita objetivo

```python
ss_earth = Orbit.from_ephem(Sun, earth, date_launch)
ss_mars = Orbit.from_ephem(Sun, mars, date_arrival)
```

Ahora podemos resolver la maniobra que nos llevará de la Tierra a Marte. Después de resolverlo, solo necesitamos aplicarlo a la órbita de salida para resolver la transferencia:

```python
# Resolver para la maniobra de transferencia
man_lambert = Maneuver.lambert(ss_earth, ss_mars)
print(man_lambert)

# Obtenemos la transferencia y las órbitas finales.
ss_trans, ss_target = ss_earth.apply_maneuver(man_lambert, intermediate=True)
print("La orbita de transferencia es: ",ss_trans)
print("La orbita final es: ",ss_target)
```

# **Ejemplo de la orbita de la Mision New Horizons de la NAsA**
Misión New Horizons de la NASA destinada a explorar Plutón, sus satélites y asteroides del cinturón de Kuiper. La sonda se lanzó desde Cabo Cañaveral el 19 de enero de 2006

```python
from astropy import time
from astropy import units as u

from poliastro import iod

from poliastro.bodies import Sun, Earth, Jupiter
from poliastro.ephem import Ephem
from poliastro.frames import Planes
from poliastro.plotting import StaticOrbitPlotter
from poliastro.twobody import Orbit
from poliastro.util import norm
```

Primero se insertó la sonda en una órbita elíptica de estacionamiento de la Tierra con una altitud de perigeo de 165 km y una altitud de apogeo de 215 km.
Nota: Apogeo es el punto en una órbita elíptica alrededor de la Tierra en la que un cuerpo se encuentra más alejado del centro de esta. El punto orbital opuesto, es el más cercano y se llama perigeo.

```python
r_p = Earth.R + 165 * u.km
r_a = Earth.R + 215 * u.km

a_parking = (r_p + r_a) / 2
ecc_parking = 1 - r_p / a_parking

parking = Orbit.from_classical(
    Earth,
    a_parking,
    ecc_parking,
    0 * u.deg,
    0 * u.deg,
    0 * u.deg,
    0 * u.deg,  # We don't mind
    time.Time("2006-01-19", scale="utc"),
)

print(parking.v)
parking.plot();
```

Salida hiperbólica
Velocidad excesiva hiperbólica:


Relación entre la velocidad orbital (v), la velocidad de escape local (Ve) y el exceso de velocidad hiperbólica (v∞):

V^2=Ve^2 +(v∞)^2


```python
C_3_A = 157.6561 * u.km ** 2 / u.s ** 2  # OpcionA: tomar como Velocidad (v∞)^2 la reportada en el perigeo para calcular Vexit

a_exit = -(Earth.k / C_3_A).to(u.km)
ecc_exit = 1 - r_p / a_exit

exit = Orbit.from_classical(
    Earth,
    a_exit,
    ecc_exit,
    0 * u.deg,
    0 * u.deg,
    0 * u.deg,
    0 * u.deg,  # We don't mind
    time.Time("2006-01-19", scale="utc"),
)

norm(exit.v).to(u.km / u.s)    #Vexit = Ve
```

Después de una corta estancia en la órbita de estacionamiento, la nave espacial fue inyectada en la órbita heliocéntrica deseada por la segunda etapa Centauro y la tercera etapa Star 48B. En el agotamiento del combustible del motor de la seccion Star 48B, la nave espacial New Horizons alcanzó la velocidad de salida más alta de la Tierra, estimada en 16,2 km / s, convirtiéndose en la nave espacial más rápida jamás lanzada desde la Tierra.

```python
v_estimated = 16.2 * u.km / u.s

print(
    "Relative error of {:.2f} %".format(
        (norm(exit.v) - v_estimated) / v_estimated * 100
    )
)
```

Entonces el error permanece dentro del mismo orden de magnitud.

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(8, 8))
op = StaticOrbitPlotter(ax=ax)

op.plot(parking)
op.plot(exit)

ax.set_xlim(-8000, 8000)
ax.set_ylim(-20000, 20000);
```

Option b): Calcular v∞ usando el sobrevuelo de Jupyter
Según Wikipedia, el acercamiento más cercano ocurrió a las 05:43:40 UTC. Podemos usar estos datos para calcular la solución del problema de Lambert entre la Tierra y Júpiter:

```python
nh_date = time.Time("2006-01-19 19:00", scale="utc").tdb
nh_flyby_date = time.Time("2007-02-28 05:43:40", scale="utc").tdb
nh_tof = nh_flyby_date - nh_date

nh_r_0, v_earth = Ephem.from_body(Earth, nh_date).rv(nh_date)
nh_r_f, v_jup = Ephem.from_body(Jupiter, nh_flyby_date).rv(nh_flyby_date)

(nh_v_0, nh_v_f), = iod.lambert(Sun.k, nh_r_0, nh_r_f, nh_tof)
```

El exceso de velocidad hiperbólica se mide con respecto a la Tierra:

```python
C_3_lambert = (norm(nh_v_0 - v_earth)).to(u.km / u.s) ** 2
C_3_lambert
```

```python
print("Error relativo de {:.2f} %".format((C_3_lambert - C_3_A) / C_3_A * 100))
```

Lo cual, nuevamente, se mantiene dentro del mismo orden de magnitud de la cifra dada al informe Guo & Farquhar.

New Horizons se aproximó a Júpiter entre febrero y marzo de 2007, para aprovechar la asistencia gravitatoria del planeta y adquirir así una diferencia de velocidad de unos 4 023 m/s (14 482 km/h).

```python
nh = Orbit.from_vectors(Sun, nh_r_0, nh_v_0, nh_date)

op = StaticOrbitPlotter(plane=Planes.EARTH_ECLIPTIC)

op.plot_body_orbit(Jupiter, nh_flyby_date)
op.plot_body_orbit(Earth, nh_date)
op.plot(nh, label="New Horizons", color="k");
```
